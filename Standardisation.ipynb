```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
from sklearn.datasets import load_boston
```

```python
data = load_boston()
```

```python
dir(data)
```

```python
data.data
```

```python
df = pd.DataFrame(data.data,columns = data.feature_names)
```

```python
df.head()
```

```python
df['target'] = data.target
```

```python
df = df[['CRIM','RM','LSTAT','target']]
```

```python
df.head()
```

```python
df.describe()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state = 42)
```

```python
x_train.shape , x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
sc.fit(x_train)
```

```python
x_train_scaled = sc.transform(x_train)
x_test_scaled = sc.transform(x_test)
```

```python
# Scaler Stores mean of each features
```

```python
sc.mean_
```

```python
# Scaler stores Std of each features
```

```python
sc.scale_
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled,columns = x_train.columns)
```

```python
x_test_scaled = pd.DataFrame(x_test_scaled,columns = x_test.columns)
```

```python
x_train_scaled.head()
```

```python
x_train_scaled.describe()
```

```python
plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
x_train['CRIM'].plot(kind = 'kde' , color = 'g', label ='Before Scalling')
plt.subplot(1,2,2)
x_train_scaled['CRIM'].plot(kind = 'kde', color ='r' , label ='Before Scaling')
plt.show()
               
```

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.set_title('Before Scaling')
sns.kdeplot(x_train['CRIM'] , ax =ax1)
sns.kdeplot(x_train['RM'],ax = ax1)
sns.kdeplot(x_train['LSTAT'],ax= ax1)

ax2.set_title('After Scalling')
sns.kdeplot(x_train_scaled['CRIM'],ax = ax2)
sns.kdeplot(x_train_scaled['RM'],ax = ax2)
sns.kdeplot(x_train_scaled['LSTAT'],ax = ax2)
plt.show()
```
